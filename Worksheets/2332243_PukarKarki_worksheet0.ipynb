```python
def length_conversion(value, conversion_type):
    """
    Convert length between meters and feet.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('m_to_ft' or 'ft_to_m').

    Returns:
    float: Converted value.
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084  # 1 meter = 3.28084 feet
    elif conversion_type == "ft_to_m":
        return value / 3.28084  # 1 foot = 0.3048 meters
    else:
        return None

def weight_conversion(value, conversion_type):
    """
    Convert weight between kilograms and pounds.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('kg_to_lbs' or 'lbs_to_kg').

    Returns:
    float: Converted value.
    """
    if conversion_type == "kg_to_lbs":
        return value * 2.20462  # 1 kg = 2.20462 lbs
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462  # 1 lb = 0.453592 kg
    else:
        return None

def volume_conversion(value, conversion_type):
    """
    Convert volume between liters and gallons.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('L_to_gal' or 'gal_to_L').

    Returns:
    float: Converted value.
    """
    if conversion_type == "L_to_gal":
        return value * 0.264172  # 1 liter = 0.264172 gallons
    elif conversion_type == "gal_to_L":
        return value / 0.264172  # 1 gallon = 3.78541 liters
    else:
        return None

def main():
    """
    Main function to prompt user input and perform conversions.
    """
    print("Welcome to Unit Converter!")
    print("Select the conversion type:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kilograms <-> pounds)")
    print("3. Volume (liters <-> gallons)")

    try:
        choice = int(input("Enter your choice (1/2/3): "))
        if choice not in [1, 2, 3]:
            print("Invalid choice! Please select a valid option.")
            return

        value = float(input("Enter the value to convert: "))

        if choice == 1:
            print("1. Convert meters to feet")
            print("2. Convert feet to meters")
            length_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "m_to_ft" if length_choice == 1 else "ft_to_m"
            result = length_conversion(value, conversion_type)

        elif choice == 2:
            print("1. Convert kilograms to pounds")
            print("2. Convert pounds to kilograms")
            weight_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "kg_to_lbs" if weight_choice == 1 else "lbs_to_kg"
            result = weight_conversion(value, conversion_type)

        elif choice == 3:
            print("1. Convert liters to gallons")
            print("2. Convert gallons to liters")
            volume_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "L_to_gal" if volume_choice == 1 else "gal_to_L"
            result = volume_conversion(value, conversion_type)

        print(f"Converted Value: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
def calculate_sum(numbers):
    """
    Calculate the sum of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Sum of numbers.
    """
    total = 0
    for num in numbers:
        total += num
    return total

def calculate_average(numbers):
    """
    Calculate the average of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Average value.
    """
    total = 0
    count = 0
    for num in numbers:
        total += num
        count += 1
    return total / count if count > 0 else 0

def calculate_maximum(numbers):
    """
    Find the maximum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Maximum number.
    """
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

def calculate_minimum(numbers):
    """
    Find the minimum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Minimum number.
    """
    min_value = numbers[0]
    for num in numbers:
        if num < min_value:
            min_value = num
    return min_value

def main():
    """
    Main function to prompt user input and perform calculations.
    """
    print("Math Operations on a List of Numbers")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Enter your choice (1/2/3/4): "))
        if choice not in [1, 2, 3, 4]:
            print("Invalid choice! Please select a valid option.")
            return

        numbers = input("Enter numbers separated by spaces: ").split()
        numbers = [float(num) for num in numbers]

        if not numbers:
            print("Error: No numbers provided.")
            return

        if choice == 1:
            result = calculate_sum(numbers)
            operation = "Sum"
        elif choice == 2:
            result = calculate_average(numbers)
            operation = "Average"
        elif choice == 3:
            result = calculate_maximum(numbers)
            operation = "Maximum"
        else:
            result = calculate_minimum(numbers)
            operation = "Minimum"

        print(f"{operation} of given numbers: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


if __name__ == "__main__":
    main()
```

```python
#1 Extract Every other element

def extract_every_other(lst):
  """
  Extracts every other element from a list, starting from the first element.

  Parameters:
  lst (list): The input list.

  Returns:
  list: A new list containing every other element from the input list.
  """
  return lst[::2]

#2 Slice a sublist

def get_sublist(lst, start, end):
  """
  Extracts a sublist from a given list based on specified start and end indices.

  Parameters:
  lst (list): The input list.
  start (int): The starting index for the sublist.
  end (int): The ending index for the sublist.

  Returns:
  list: A new list containing the elements from the input list within the specified
  """
  return lst[start:end+1]

#3 Reverse a List using slicing

def reverse_list(lst):
  """
  Reverses a list using slicing.

  Parameters:
  lst (list): The input list.

  Return:
  List: a reversed List
  """
  return lst[::-1]

#4 Remove the first and last element

def remove_first_last(lst):
  """
  Removes the first and last elements of a list and returns the resulting sublist.

  Parameters:
  lst(list): The input list.

  Returns:
  list: A new list containing all elements except the first and last elements of the input list.
  """
  return lst[1:-1]

#5. Get the first n element

def get_first_n(lst, n):
  """
  Extracts the first n elements from a list.

  Parameters:
  lst (list): The input list.
  n (int): Number of elements to extract.

  Returns:
  list: A new list containing the first n elements from the input list.
  """
  return lst[:n]

#6. Extract Element from the end

def get_last_n(lst, n):
  """
  Extracts the last n elements from a list using slicing.

  Parameters:
  lst (list): The input list.
  n (int): Number of elements to extract from the end.

  Returns:
  list: A new list containing the last n elements.
  """
  return lst[-n:]

#7. Extract Element in reverse order

def reverse_skip(lst):
  """
  Extracts elements in reverse order, skipping one in between, starting from the second-to-last element.

  Parameters:
  lst (list): The input list.

  Returns:
  list: A new list containing elements in reverse order, skipping one in between, starting from the second-to-last element.
  """
  return lst[-2::-2]

#Example list
sample_list =  [1,2,3,4,5,6]

print("1. Extract every other element:", extract_every_other(sample_list))
print("2. Sublist (2 to 4):", get_sublist(sample_list, 2, 4))
print("3. Reversed list:", reverse_list(sample_list))
print("4. Without first & last:", remove_first_last(sample_list))
print("5. First 3 elements:", get_first_n(sample_list, 3))
print("6. Last 2 elements:", get_last_n(sample_list, 2))
print("7. Reverse skip:", reverse_skip(sample_list))
```

```python
#1. flatten a Nested List

def flatten(lst):
  """
  Flattens a nested list into a single list.

  Parameters:
  lst (list): The input nested list.

  Returns:
  list: A new list containing all the elements from the nested list.
  """
  result = []
  for sublist in lst:
    if isinstance(sublist, list):
      result.extend(flatten(sublist))
    else:
      result.append(sublist)
  return result

#2. Accessing Nested List Elements
def access_nested_element(lst, indices):
  """
  Extracts a specific element from a nested list given its indices.

  Parameters:
  lst (list): The input nested list.
  indices (list): The indices specifying the path to the element.

  Return:
  The extracted element.
  """
  element = lst
  for index in indices:
    element = element[index]
  return element

#3. Sum of All Elements in a Nested List

def sum_nested(lst):
  """
  Calculates the sum of all numbers in a nested list, regardless of depth.

  Parameters:
  lst (list): The input nested list.

  Returns:
  int/float: The sum of all elements.
  """
  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested(item)
    else:
      total +=item
  return total

# 4. Remove Specific Element from a Nested List

def remove_element(lst, elem):
  """
  Removes all occurrences of a specific element from a nested list.

  Parameters:
  lst (list): The input nested list.
  elem: The element to remove.

  Returns:
  list: A new list with all occurrences of the specified element removed.
  """
  result = []
  for item in lst:
    if isinstance(item, list):
      filtered_list = remove_element(item, elem)
      if filtered_list:
        result.append(filtered_list)
    elif item != elem:
      result.append(item)
  return result

# 5. Find the Maximum Element in a Nested List

def find_max(lst):
  """
  Finds the maximum element in a nested list.

  Parameters:
  lst (list): The input nested list.

  Returns:
  The maximum element.
  """
  max_value = float('-inf')
  for item in lst:
    if isinstance(item, list):
      max_value = max(max_value, find_max(item))
    else:
      max_value = max(max_value, item)
  return max_value

#6. Count Occurrences of an Element in a Nested List

def count_occurrences(lst, elem):
  """
  Counts the occurrences of a specific element in a nested list.

  Parameters:
  lst (list): The input nested list.
  elem: The element to count.

  Returns:
  int: The number of occurrences of elem.
  """
  count = 0
  for item in lst:
    if isinstance(item, list):
      count += count_occurrences(item, elem)
    elif item == elem:
      count +=1
  return count

#7. Flatten a List of Lists of Lists

def deep_flatten(lst):
  """
  Flattens a deeply nested list into a single list.

  Parameters:
  lst (list): The input deeply nested list.
  Returns:
  list: A new list containing all the elements from the deeply nested list.
  """
  result = []
  for item in lst:
    if isinstance(item, list):
      result.extend(deep_flatten(item))
    else:
      result.append(item)
  return result

# 8. Nested List Average

def average_nested(lst):
  """
   Calculates the average of all elements in a nested list.

   Parameters:
   lst (list): The input nested list.

   Returns:
   float: The average of all elements.
  """
  flat_list = deep_flatten(lst)
  return sum(flat_list) / len(flat_list) if flat_list else 0


#Example Input list
nested_list_1 = [[1,2],[3,4],[5]]
nested_list_2 = [[1,2,3],[4,5,6],[7,8,9]]
nested_list_3 = [[1,2],[3,[4,5]],6]
nested_list_4 = [[1,2],[3,2],[4,5]]
nested_list_5 = [[[1,2],[3,4]],[[5,6],[7,8]]]


print("1. Flatten a Nested List: ", flatten(nested_list_1))
print("2. Access Nested Element (index [1,2]): ", access_nested_element(nested_list_2, [1,2]))
print("3. Sum of All Elements:", sum_nested(nested_list_3))
print("4. Remove Element (elem=2):", remove_element(nested_list_4, 2))
print("5. Max Element in Nested List:", find_max(nested_list_3))
print("6. Count Occurrences (elem=2):", count_occurrences(nested_list_4, 2))
print("7. Deep Flatten:", deep_flatten(nested_list_5))
print("8. Average of Nested List:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("A * A⁻¹ = Identity Matrix:\n", identity_check)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nMatrix AB:\n", AB)
print("\nMatrix BA:\n", BA)
if not np.array_equal(AB, BA):
    print("\nAB is NOT equal to BA")

# 3. Prove (AB)ᵀ = Bᵀ * Aᵀ
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)ᵀ:\n", AB_transpose)
print("\nBᵀ * Aᵀ:\n", BT_AT)
if np.array_equal(AB_transpose, BT_AT):
    print("\n(AB)ᵀ is equal to Bᵀ * Aᵀ")

# 4. Solve Linear Equations using Inverse Methods
# System of equations:
#  2x - 3y + z = -1
#  x - y + 2z = -3
#  3x + y - z = 9

A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_matrix = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A_matrix)
solution = np.dot(A_inv, B_matrix)
print("\nSolution for Linear Equations (x, y, z):\n", solution)
```
